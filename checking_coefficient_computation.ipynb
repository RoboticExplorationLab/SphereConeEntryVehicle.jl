# Comparing Aerodynamics Coefficients With Original Paper 
Analytic Hypersonic Aerodynamics for Conceptual Design of Entry Vehicles

```julia
using ChebyshevApprox
using ForwardDiff
using LinearAlgebra
include("AerodynamicsCoefficients.jl")
```

### Tests for the cone 
We select a full cone of angle δ = 5 deg. We test it with an angle of attack α = 20 deg. 

```julia
#Define the geometry
r_min = 0.0
r_max = 1.0
A_ref = pi * r_max^2
α = 20.0/ 360 * 2 * pi;
```

```julia
#List of cone angles
δ_vect = [5.0, 15.0, 30.0] / 360 * 2 * pi

for δ in δ_vect
    l = r_max / tan(δ)
    C_F2, C_τ2 = aerodynamics_cone_coefficents(α, r_max, r_min, l) 
    normalized_C_F2 = C_F2 / A_ref
    normalized_C_τ2 = C_τ2 / A_ref / l
    println("normalized C_X2", normalized_C_F2[1])
    println("normalized C_Y2", normalized_C_F2[2])
    println("normalized C_n2", normalized_C_τ2[3])
end
```

### Comparison with the original paper.
The points at alpha=0 deg in the paper corresponds to alpha=20 deg in our formulation (because of beta).
We see that the values for C_A = -C_X are matching for delta in [5, 15, 30] deg. 
We see that the values for C_S = C_Y are matching for delta in [5, 15, 30] deg.

### Tests for the sphere

```julia
#List of cone angles
δ_vect = [5.0, 15.0, 30.0] / 360 * 2 * pi

for δ in δ_vect
    r_max = l * tan(δ)
    C_F2, C_τ2 = aerodynamics_sphere_coefficents(α, r_max, r_min, l) 
    normalized_C_F2 = C_F2 / A_ref
    println("normalized C_X2", normalized_C_F2[1])
    println("normalized C_Y2", normalized_C_F2[2])
end
```
